By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10 001st prime number?


```python
import math
def is_prime(n):
  for i in range(2,int(math.sqrt(n))+1):
    if (n%i) == 0:
      return False
  return True

search = 10001

res = None
count = 0
step = 2
while(not res):
    # if step % 10 ==0:
    #     print(step)
    #     print(count)

    if is_prime(step):
        count += 1
        print(f"count {count} prime {step}")
        
        if count == search:
            res = step
    step += 1

step-1
```
